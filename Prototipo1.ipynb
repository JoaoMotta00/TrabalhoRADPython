```python
import sqlite3

#Cria o banco de dados
connection = sqlite3.connect("alunos.db")
alunos = connection.cursor()

alunos.execute("""
    CREATE TABLE IF NOT EXISTS alunos(matricula INTEGER,
                                     cpf INTEGER,
                                     nome TEXT NOT NULL,
                                     data_nascimento TEXT NOT NULL,
                                     endereco TEXT NOT NULL,
                                     PRIMARY KEY(matricula, cpf))
""")
```

```python
#Outras operações da tabela INSERT, DELETE, UPDATE, SELECT

opcao = 0


while opcao != 5 :
    print("------------------------------")
    print("1 - INSERIR DADO NA TABELA")
    print("2 - ATUALIZAR DADO NA TABELA")
    print("3 - DELETAR DADO NA TABELA")
    print("4 - EXIBIR DADOS DA TABELA")
    print("5 - ENCERRAR O PROGRAMA")
    print("------------------------------")
    opcao = int(input("SELECIONE UMA OPÇÃO: "))
    
    if opcao == 1:
        #INSERT
        matricula = int(input("DIGITE A MATRICULA DO ALUNO: "))
        cpf = int(input("DIGITE O CPF DO ALUNO: "))
        nome = input("DIGITE O NOME DO ALUNO: ")
        data = input("DIGITE A DATA DE NASCIMENTO DO ALUNO: ")
        endereco = input("DIGITE O ENDEREÇO DO ALUNO: ")
        
        alunos.execute("INSERT INTO alunos(matricula, cpf, nome, data_nascimento, endereco) VALUES(?, ?, ?, ?, ?)", 
                       (matricula,
                       cpf,
                       nome,
                       data,
                       endereco))
        
        connection.commit()
    
    elif opcao == 2:
        #UPDATE

        seletorAluno = int(input("DIGITE A MATRICULA DO ALUNO PARA EDITAR"))
        print("1 - NOME")
        print("2 - ENDEREÇO")
        seletor = int(input("SELECIONE O QUE QUER EDITAR"))

        if seletor == 1:
            nome = input("DIGITE O NOVO NOME: ")
            alunos.execute("UPDATE alunos SET nome = ? WHERE matricula = ?", (nome, seletorAluno))
            connection.commit()

        elif seletor == 2:
            endereco = input("DIGITE O NOVO ENDEREÇO: ")
            alunos.execute("UPDATE alunos SET endereco = ? WHERE matricula = ?", (endereco, seletorAluno))
            connection.commit()
    
    elif opcao == 3:
        #DELETE
        matricula = int(input("DIGITE A MATRÍCULA QUE DESEJA APAGAR: "))

        alunos.execute("DELETE FROM alunos WHERE matricula = ?", (matricula,))

        connection.commit()

    elif opcao == 4:
        #SELECT
        for i in alunos.execute("SELECT rowid, * FROM alunos ORDER BY nome"):
            print(i)

connection.close()

```
